# Aggregating the PBM data from the many files it is in
### In this script, I will aggregate the many files of the PBM experiment into a single csv for easy use. Only non-gapped 8-mer sequences will be treated here.

```python
# Useful CONSTS:
DATA_TF_PATH = "/home/labs/davidgo/ortalh/TF_analysis_data/data_TF_for_analysis"
MY_DATA_DIR = "/home/labs/davidgo/matanyaw/data"
```

```python
TF_list = [tf for tf in os.listdir(DATA_TF_PATH) if os.path.isdir(os.path.join(DATA_TF_PATH, tf))]
TF_list
```

```python
### This is ortal's code to extract the TF data from the files
print("##### Extracting txt files from TF path #####")
print("##### Reading TF files #####")
TF_data_files = ortals.filter_TF_txt_files(ortals.get_all_txt_in_folder(DATA_TF_PATH))
```

```python
pdb_data, _ = um.parse_8mer_tf_file(TF_data_files[0])
for i in range(1, len(TF_data_files)):
# for i in range(1, 3):
    df, _ = um.parse_8mer_tf_file(TF_data_files[i])
    if df is None:
        continue
    pdb_data = pdb_data.merge(df, on=["8-mer", "reverse_complement"], how="inner", suffixes=("", ""))
    print(i, pdb_data.shape)
```

```python
pdb_data
```

```python
def convert_df_into_multiindex(df: pd.DataFrame, id_cols=["8-mer", "reverse_complement"]):
    """
    Converts a DataFrame into a MultiIndex DataFrame.

    Args:
        df (pd.DataFrame): The DataFrame to convert.
        id_cols (list of str, optional): The column names to use as the primary index.
            Defaults to ["8-mer", "reverse_complement"].

    Returns:
        pd.DataFrame: A new DataFrame with a MultiIndex column structure.
    """
    multi_columns = []
    for col in df.columns:
        if col in id_cols:
            multi_columns.append(("ID", col))
        else:
            parts = col.split("_", 1)
            if len(parts) == 2:
                category, tf_name = parts
            else:
                category, tf_name = col, "unkown"
            multi_columns.append((category, tf_name))
    new_df = df.copy()
    new_df.columns = pd.MultiIndex.from_tuples(multi_columns)
    return new_df

```

```python
pbm_indexed_df = pdb_data.set_index("8-mer", inplace=False)
pbm_indexed_df.drop("reverse_complement", axis=1, inplace=True)
pbm_indexed_df
```

```python

pbm_multicol_df = convert_df_into_multiindex(pbm_indexed_df, id_cols=[])
pbm_multicol_df.sort_index(axis=1,level=0, inplace=True)
pbm_multicol_df
```

```python
pbm_multicol_df.to_csv(os.path.join(MY_DATA_DIR, "pbm_8mer_aggregated_data.csv"))
```

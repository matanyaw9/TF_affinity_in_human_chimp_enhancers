Let's see whitch TFs bind to which areas in the Derived and Ancestral DNA Sequences:

```python
START_INDEX = 0
END_INDEX = 5
WINDOW_SIZE = 8
OUTPUT_DIR = "/home/labs/davidgo/matanyaw/data/DRAFT_tf_2_locus_zscores/"
if os.path.exists(OUTPUT_DIR):
    os.system(f"rm -r {OUTPUT_DIR}")
os.makedirs(OUTPUT_DIR)
```

```python

for i, row in mpra_df.iloc[START_INDEX:END_INDEX, :].iterrows():
    # Get the ancestral and derived DNA sequences
    anc_seq = row['sequence_ancestral'].upper()
    der_seq = row['sequence_derived'].upper()
    
    # Break the sequences into 8-mer windows
    anc_seq_windows = um.get_running_window_numpy(seq=anc_seq, window_size=WINDOW_SIZE)
    der_seq_windows = um.get_running_window_numpy(seq=der_seq, window_size=WINDOW_SIZE)
    
    # Choose for each 8-mer either itself or its reverse complement
    anc_min_windows = um.get_min_seq(anc_seq_windows)
    der_min_windows = um.get_min_seq(der_seq_windows)

    # Get the z-scores for the 8-mers and TFs
    locus = row['oligo'].replace(":", "_")
    anc_pbm_df_unfiltered = zscore_df.loc[anc_min_windows]
    der_pbm_df_unfiltered = zscore_df.loc[der_min_windows]
    # Get the e-scores for the 8-mers and TFs
    anc_escore_df = escore_df.loc[anc_min_windows] > 0.35
    der_escore_df = escore_df.loc[der_min_windows] > 0.35

    
    # This is to have the original sequence and not the reverse complement
    anc_pbm_df_unfiltered.index = anc_seq_windows
    anc_escore_df.index = anc_seq_windows
    der_pbm_df_unfiltered.index = der_seq_windows
    der_escore_df.index = der_seq_windows


    anc_pbm_df = anc_pbm_df_unfiltered.where(anc_escore_df, 0)
    der_pbm_df = der_pbm_df_unfiltered.where(der_escore_df, 0)


    # Save the TF z-scores for the two versions of the locus
    locus_dir = os.path.join(OUTPUT_DIR, locus)
    if not os.path.exists(locus_dir):
        os.makedirs(locus_dir)
    
    anc_pbm_df_T = anc_pbm_df.T
    der_pbm_df_T = der_pbm_df.T
    anc_pbm_df_T.to_csv(os.path.join(locus_dir, "Ancestral_PBM.csv" ))
    der_pbm_df_T.to_csv(os.path.join(locus_dir, "Derived_PBM.csv" ))

    anc_pbm_df_T.columns = range(anc_pbm_df_T.shape[1])
    der_pbm_df_T.columns = range(der_pbm_df_T.shape[1])

    diff_pbm_df_T = der_pbm_df_unfiltered - anc_pbm_df_unfiltered
    diff_pbm_df_T = diff_pbm_df_T.where(anc_escore_df | der_escore_df, 0).T

    # Filter on E-score. E-score must be above 0.35 for at least on of the sequences to be considered




    diff_pbm_df_T.to_csv(os.path.join(locus_dir, "Difference_PBM.csv" ))

    um.create_heatmap(anc_pbm_df_T, title="TF Affinity in Ancestral Sequence", locus=locus, path=locus_dir)
    um.create_heatmap(der_pbm_df_T, title="TF Affinity in Derived Sequence", locus=locus, path=locus_dir)
    um.create_heatmap(diff_pbm_df_T, title="TF Affinity Difference: Derived - Ancestral", locus=locus, path=locus_dir)

    print(f"Done with {i+1}/{mpra_df.shape[0]}")
```

Now let us create a df of all TF and all seqs to see the max gaps

```python
diff_pbm_file_paths = []

for root, dirs, files in os.walk(os.path.join(MY_DATA_DIR, "tf_to_loci_files_new")):
    for file in files:
        if file == "Difference_PBM.csv":
            diff_pbm_file_paths.append(os.path.join(root, file))

len(diff_pbm_file_paths)
```

```python
TF_binding_stronger_in_derived = pd.DataFrame()
TF_binding_stronger_in_ancestral = pd.DataFrame()

for file in diff_pbm_file_paths:
    df = pd.read_csv(file, index_col=0)
    locus = file.split("/")[-2]
    max_pos_diff_per_tf = df.max(axis=1)
    max_neg_diff_per_tf = df.min(axis=1)
    max_pos_diff_per_tf.name = locus
    max_neg_diff_per_tf.name = locus
    TF_binding_stronger_in_derived = pd.concat([TF_binding_stronger_in_derived, max_pos_diff_per_tf], axis=1)
    TF_binding_stronger_in_ancestral = pd.concat([TF_binding_stronger_in_ancestral, max_neg_diff_per_tf], axis=1)

TF_binding_stronger_in_derived.to_csv(os.path.join(MY_DATA_DIR, "TF_binding_stronger_in_derived_all_loci.csv"))
TF_binding_stronger_in_ancestral.to_csv(os.path.join(MY_DATA_DIR, "TF_binding_stronger_in_ancestral_all_loci.csv"))

TF_binding_stronger_in_derived
```

```python
from matplotlib.colors import CenteredNorm
import seaborn as sns

norm = CenteredNorm(vcenter=0)  

sns.heatmap(TF_binding_stronger_in_derived, cmap='RdBu_r', annot=False, norm=norm)

plt.title("TF_binding_stronger_in_derived_all_loci".replace("_", " "), fontsize=16, pad=20)

# Axis labels
plt.xlabel("Loci", fontsize=12)
plt.ylabel("Transcription Factor", fontsize=12)

# Ensure path exists if provided
save_path = os.path.join("TF_binding_stronger_in_derived_all_loci" + ".png")
plt.savefig(save_path)
plt.show()
plt.close()
```

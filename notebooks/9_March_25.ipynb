Now let's create a df for a locus


```python
def create_df_for_locus(anc_str, der_str, zscore_df):
    if len(anc_str) != len(der_str):
        raise ValueError("anc and der must be of the same length")
    anc_np, der_np = np.array(list(anc_str)), np.array(list(der_str))

    return np.where(der_np != anc_np)

```
